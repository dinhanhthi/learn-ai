```python
import pandas as pd
df_ds = pd.read_csv("./Topic_Survey_Assignment.csv", index_col=0)
df_ds
```

## Problem 2

```python
df_ds_sorted = df_ds.sort_values(by=["Very interested"], ascending=False)
df_ds_sorted
```

```python
df_ds_percent = round(df_ds_sorted/2233,4)
df_ds_percent
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(111)

n_groups = 6
ind = np.arange(n_groups)
width = 0.25

plt.bar(ind, df_ds_percent['Very interested'], width, label='Very interested', color='#5cb85c')
plt.bar(ind + width, df_ds_percent['Somewhat interested'], width, label='Somewhat interested', color='#5bc0de')
plt.bar(ind + 2*width, df_ds_percent['Not interested'], width, label='Not interested', color='#d9534f')

plt.xticks(ind + width, ('Data Analysis / Statistics', 'Machine Learning',
                         'Data Visualization', 'Big Data (Spark / Hadoop)', 'Deep Learning', 'Data Journalism'),
                        rotation='vertical')
plt.legend(loc='best', prop={'size':14})
ax.tick_params(axis='both', which='major', labelsize=14)
plt.title("Percentage of Respondents' Interest in Data Science Areas", fontsize=16)
plt.tick_params(top=False, left=False, right=False, labelleft=False, labelbottom=True)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)

def show_percentage(lst, pos):
    for i, v in enumerate(lst):
        ax.text(pos + 4*i*width, v + 0.02, str(round(v*100,2)), color='gray', fontsize=14)
        
show_percentage(df_ds_percent['Very interested'], -0.1)
show_percentage(df_ds_percent['Somewhat interested'], -0.1 + width)
show_percentage(df_ds_percent['Not interested'], -0.1 + 2*width)

plt.show()
```

## Problem 3

```python
df_crime = pd.read_csv("./Police_Department_Incidents_-_Previous_Year__2016_.csv")
```

```python
df_crime
```

```python
df_nb_crime = df_crime.groupby(['PdDistrict'], as_index=False).size().reset_index(name='Count')
df_nb_crime.rename(columns={"PdDistrict": "Neighborhood"}, inplace=True)
df_nb_crime
```

## Question 4

```python
import folium
```

```python
sf_geo = r'san-francisco.geojson' # geojson file

# create a plain SF map
sanfran_map = folium.Map(location=[37.77, -122.42], zoom_start=12)
sanfran_map
```

```python
sanfran_map.choropleth(
    geo_data=sf_geo,
    data=df_nb_crime,
    columns=['Neighborhood', 'Count'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Rate in San Francisco'
)
sanfran_map
```
